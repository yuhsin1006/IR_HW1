```python
#---------------------- 讀 Query的 List ------------------------------------

#querylist = open('C:/Users/Ray/Desktop/資訊檢索/Homework1/query_list.txt','r')
#print(querylist.read())

## Open file
fp = open('C:/Users/Ray/Desktop/資訊檢索/Homework1/query_list.txt','r')
ql = [ ]
line = fp.readline()
#querylist.append(line)

## 用 while 逐行讀取檔案內容，直至檔案結尾
while line:
 #   print (line)
    line = line.replace('\n','')
    ql.append(line)
    line = fp.readline()
 
fp.close()
#print (ql)


#---------------------- 讀 DOC 的 List ----------------------------------------

#doclist = open('C:/Users/Ray/Desktop/資訊檢索/Homework1/doc_list.txt','r')
#print(doclist.read())

## Open file
fp = open('C:/Users/Ray/Desktop/資訊檢索/Homework1/doc_list.txt','r')
docl = [ ]
line = fp.readline()
#querylist.append(line)

## 用 while 逐行讀取檔案內容，直至檔案結尾
while line:
 #   print (line)
    line = line.replace('\n','')
    docl.append(line)
    line = fp.readline()
 
fp.close()
#print (docl)
```

```python
#------------------------- 讀且取每個 Query 內的字 ----------------------------

# Open file
qryl = 0
querylist = [ ]

while qryl < len(ql) :
    fp = open('C:/Users/Ray/Desktop/資訊檢索/Homework1/Query/'+ql[qryl],'r')
    qlt = fp.read()
    qlt = qlt.replace('-1','')
    qlt = qlt.split()
    querylist = querylist + qlt
    fp.close()
    qryl += 1

    
#print(total)
querylist = list(set(querylist))

voc = querylist

print(voc)
print(len(voc))
```

```python
#------------------------ 讀且取每個 Doc 內的字 --------------------------------

# Open file
dcl = 0
doc = [ ]

#total = querylist


while dcl < len(docl) :
    fp = open('C:/Users/Ray/Desktop/資訊檢索/Homework1/Document/'+docl[dcl],'r')
    i = 0
    while i < 3:
        buffer = fp.readline()
        i += 1
    
    doc = fp.read()
    doc = doc.replace('-1','')
    doc = doc.split()
    voc = voc + doc
    voc = list(set(voc))
    
    fp.close()
    dcl += 1
    
print(voc)
print(len(voc))
```
